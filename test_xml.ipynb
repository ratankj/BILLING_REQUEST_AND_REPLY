```python
import os
import xml.etree.ElementTree as et
import pandas as pd

# Define the directory containing the XML files
path = r"C:\Users\Ratan Kumar Jha\Desktop\BILLING_REQUEST_AND_REPLY\Reply\Reply"

# Initialize lists to store data
Consumer_ID_list = []
Meter_Number_list = []
Installation_Number_list = []
Area_Type_list = []
Max_Demand_KW_list = []
Max_Demand_KVA_list = []
Power_Fact_list = []
Import_KWH_list = []
Export_KWH_list = []
Export_KVAH_list = []
TOD_list = []
Read_Date_list = []
Meter_Note_list = []
MRU_list = []
MR_Schedule_Date_list = []

# Iterate through each file in the directory
for file in os.listdir(path):
    if file.endswith(".xml"):
        file_path = os.path.join(path, file)
        #print(f"Processing file: {file_path}")

        try:
            # Parse the XML file
            tree = et.parse(file_path)
            root = tree.getroot()

            # Iterate through each Data element in the XML
            for data in root.findall('.//Data'):
                Consumer_ID = data.find('Consumer_ID')
                if Consumer_ID is not None:
                    Consumer_ID_list.append(Consumer_ID.text)
                else:
                    Consumer_ID_list.append(None)

                Meter_Number = data.find('Meter_Number')
                if Meter_Number is not None:
                    Meter_Number_list.append(Meter_Number.text)
                else:
                    Meter_Number_list.append(None)

                Installation_Number = data.find('Installation_Number')
                if Installation_Number is not None:
                    Installation_Number_list.append(Installation_Number.text)
                else:
                    Installation_Number_list.append(None)

                Area_Type = data.find('Area_Type')
                if Area_Type is not None:
                    Area_Type_list.append(Area_Type.text)
                else:
                    Area_Type_list.append(None)

                Max_Demand_KW = data.find('Max_Demand_KW')
                if Max_Demand_KW is not None:
                    Max_Demand_KW_list.append(Max_Demand_KW.text)
                else:
                    Max_Demand_KW_list.append(None)

                Max_Demand_KVA = data.find('Max_Demand_KVA')
                if Max_Demand_KVA is not None:
                    Max_Demand_KVA_list.append(Max_Demand_KVA.text)
                else:
                    Max_Demand_KVA_list.append(None)

                Power_Fact = data.find('Power_Fact')
                if Power_Fact is not None:
                    Power_Fact_list.append(Power_Fact.text)
                else:
                    Power_Fact_list.append(None)

                Import_KWH = data.find('Import_KWH')
                if Import_KWH is not None:
                    Import_KWH_list.append(Import_KWH.text)
                else:
                    Import_KWH_list.append(None)

                Export_KWH = data.find('Export_KWH')
                if Export_KWH is not None:
                    Export_KWH_list.append(Export_KWH.text)
                else:
                    Export_KWH_list.append(None)

                Export_KVAH = data.find('Export_KVAH')
                if Export_KVAH is not None:
                    Export_KVAH_list.append(Export_KVAH.text)
                else:
                    Export_KVAH_list.append(None)

                TOD = data.find('TOD')
                if TOD is not None:
                    TOD_list.append(TOD.text)
                else:
                    TOD_list.append(None)

                Read_Date = data.find('Read_Date')
                if Read_Date is not None:
                    Read_Date_list.append(Read_Date.text)
                else:
                    Read_Date_list.append(None)

                Meter_Note = data.find('Meter_Note')
                if Meter_Note is not None:
                    Meter_Note_list.append(Meter_Note.text)
                else:
                    Meter_Note_list.append(None)

                MRU = data.find('MRU')
                if MRU is not None:
                    MRU_list.append(MRU.text)
                else:
                    MRU_list.append(None)

                MR_Schedule_Date = data.find('MR_Schedule_Date')
                if MR_Schedule_Date is not None:
                    MR_Schedule_Date_list.append(MR_Schedule_Date.text)
                else:
                    MR_Schedule_Date_list.append(None)

        except et.ParseError as e:
            print(f"Error parsing {file_path}: {e}")
        except Exception as e:
            print(f"Unexpected error processing {file_path}: {e}")

# Create DataFrame from the collected data
dataframe = {
    'Consumer_ID': Consumer_ID_list,
    'Meter_Number': Meter_Number_list,
    'Installation_Number': Installation_Number_list,
    'Area_Type': Area_Type_list,
    'Max_Demand_KW': Max_Demand_KW_list,
    'Max_Demand_KVA': Max_Demand_KVA_list,
    'Power_Fact': Power_Fact_list,
    'Import_KWH': Import_KWH_list,
    'Export_KWH': Export_KWH_list,
    'Export_KVAH': Export_KVAH_list,
    'TOD': TOD_list,
    'Read_Date': Read_Date_list,
    'Meter_Note': Meter_Note_list,
    'MRU': MRU_list,
    'MR_Schedule_Date': MR_Schedule_Date_list
}

reply_Data = pd.DataFrame(dataframe)

# Print the resulting DataFrame for verification
print(reply_Data)
```
